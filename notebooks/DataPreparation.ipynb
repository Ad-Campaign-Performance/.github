```python
import pandas as pd
import seaborn as sns
import os, sys
```

```python
df = pd.read_csv('data/AdSmartABdata.csv')
```

```python
df.head(10)
```

```python
yeses = list(df[["yes", "no"]].values[:,0])
nos = list(df[["yes", "no"]].values[:,1])
```

```python
response = []
for i in range(len(yeses)):
    if(yeses[i] == 1):
        response.append(1)
    elif (nos[i] == 1):
        response.append(0)
    else:
        response.append(2)
        
```

```python
len(response)
```

```python
df['response'] = response
df.head(100)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
le.fit(df['experiment'])
```

```python
le.classes_
```

```python
ex = le.transform(df['experiment'])
ex
```

```python
df['experiment'] = ex
```

```python
df.head(100)
```

```python
le.fit(df['device_make'])
le.classes_
df['device_make'] = le.transform(df['device_make'])
```

```python
df.head(100)
```

```python
le.fit(df['platform_os'])
df['platform_os'] = le.transform(df['platform_os'])

le.classes_
```

```python
df.head(100)
```

```python
le.fit(df['browser'])
df['browser'] = le.transform(df['browser'])

le.classes_
```
